Aim:
To study and implement different techniques for handling missing values in a Pandas
DataFrame using functions such as dropna(), fillna(), replace(), isnull(), notnull(), and
forward/backward filling methods, in order to improve data quality and reliability for data
analysis and machine learning applications.


Description:
Missing data is always a problem in real life scenarios. Areas like machine learning and data
mining face severe issues in the accuracy of their model predictions because of poor quality of
data caused by missing values. In these areas, missing value treatment is a major point of focus
to make their models more accurate and valid.
Handling Missing Values (Say NA Or Nan) Using Pandas - DataFrame.
1. Drop missing values using dropna() function
2. Fill the missing values using fillna() function
3. Replacing missing values using replace() with a scalar value is equivalent behavior of the
fillna() function.
4. Through isnull() and notnull() functions, we can identify the NaN as Boolean result.
5. Fill the missing values from Forward and Backward values through pad/ffill and bfill/backfill
functions respectively.

```python
import pandas as pd
import numpy as np
#index is basically row labels and colums are colum labels
df=pd.DataFrame(np.random.randn(4,3), index=['a','c','d','f'],
columns=['one','two','three'])
df = df.reindex(['a','b','c','d','e','f'])
print("Original Dataframe with NaN\n",df)
print("\n")
print("Dropped DataFrame\n",df.dropna())
print("\n")
```

```python
df=pd.DataFrame(np.random.randn(4,3), index=['a','c','d','f'],
columns=['one','two','three'])
df = df.reindex(['a','b','c'])
print("Original Dataframe:\n",df)
print("\n")
print("NaN replaced with '69.69':\n")
print(df.fillna(69.0))
```

```python
df=pd.DataFrame({'one':[10,20,80,40,50],
'two':[60,70,80,0,10]})
print("Original Dataframe:\n",df)
print("\n DataFrame with replaced values:\n")
print(df.replace({10:5,80:3}))
```

```python
df=pd.DataFrame(np.random.randn(4,3), index=['a','c','d','f'],
columns=['one','two','three'])
df = df.reindex(['a','b','c','d','e','f'])
print("Original Dataframe:\n",df)
print("\nNan-with True fill:\n")
print(df['one'].isnull())
```

```python
df=pd.DataFrame(np.random.randn(4,3), index=['a','c','d','f'],
columns=['one','two','three'])
df = df.reindex(['a','b','c','d','e','f'])
print("Original Dataframe:\n",df)
print("\nNaN - Filled with Background values:\n")
print(df.fillna(method='bfill'))
```
