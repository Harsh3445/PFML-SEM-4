Aim:
To do other matplot options.
Description:
1. Draw bar chart
2. Draw multi-bar chart with title x and y labels
3. Draw horizontal bar
4. Use color and width options
5. Draw histogram
6. Draw pie chart
-without label
- with label
-with explode

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([5, 7, 2, 15])

plt.bar(x,y)
plt.show()
```

```python
plt.bar([0.5,2.5,4.5,6.5,8.5],[50,40,70,80,20],
label="BMW",width=1)
plt.bar([1.5,3.5,5.5,7.5,9.5], [80,20,20,50,60],
label="Audi", color='c', width=1)
plt.legend()

plt.xlabel('Days')
plt.ylabel('Distance (kms)')
plt.title('Information')
plt.show()
```

```python
#Horizontal bar
plt.barh(x,y)
plt.show()
```

```python

#Color option
plt.bar(x,y, color='red')
plt.show()

#width option
plt.bar(x,y, color='g', width=0.1)
plt.show()

#Histogram
x = np.random.normal(300, 20, 250)
plt.hist(x)
plt.show()

#Pie chart
y = np.array([40, 30, 20, 10])
plt.pie(y)
plt.show()

#With Labels
y = np.array([35, 25, 25, 15])
mylabels = ["MPMC", "PPML", "ContolSystem", "PSA"]
plt.pie(y, labels = mylabels)
plt.show()

#Explode
y = np.array([35, 25, 25, 15])
mylabels = ["MPMC", "PPML", "ContolSystem", "PSA"]

myexplode = [0, 0, 0.3, 0]
plt.pie(y,labels=mylabels,explode=myexplode)
plt.show()
```

Aim:
To implement other Data Visualization tools/packages
Description:
1. Using seaborn package to do lineplot, scatterplot and barplot
2. Draw 3-D plot
3. Using plotly package to do scatterplot


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import plotly.express as px
#seaborn package
Xpoints=np.array([0,20,40,60,80,120,200])
Ypoints=np.array([10,80,100,120,150,180,250])
plt.plot(Xpoints,Ypoints)
```

```python
sns.scatterplot(x=Xpoints, y=Ypoints)
plt.show()

sns.barplot(x=Xpoints, y=Ypoints)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Data
X = np.array([1, 2, 6, 8, 12, 14, 20])
Y = np.array([3, 8, 1, 10, 12, 16, 18])
Z = np.array([5, 7, 9, 11, 13, 15, 17]) # You NEED a Z axis for 3D
# Create figure
fig = plt.figure(figsize=(12, 8))
# Create 3D axes
ax = plt.axes(projection='3d')
# 3D scatter plot
plt_3d = ax.scatter3D(X, Y, Z)
# Color bar
fig.colorbar(plt_3d)
# Show plot
plt.show()
```
