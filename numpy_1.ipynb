Aim:
To understand and implement basic NumPy array operations including array creation,
manipulation, reshaping, property inspection, and data type conversion.

```python
import numpy as np
a = np.array([[1, 2, 4], [5, 8, 7]])
print ("Array created using passed list:\n", a)
```

```python
# Creating a 3X4 array with all zeros
b = np.zeros((3, 4))
print ("\nAn array initialized with all zeros:\n", b)
```

```python
#Create a constant value array of complex type
c = np.full((3, 3), 6)
print ("\nAn array initialized with all 6s.\n", c)
```

```python
#Create an array with random values
d = np.random.random((2, 2))
print ("\nA random array:\n", d)
```

```python
# Create a sequence of integers from 0 to 30 with steps of 5
e = np.arange(0, 30, 5)
print ("\n A sequential array with steps of 5:\n", e)
```

```python
arr = np.array([[1, 2, 3, 4], [5, 2, 4, 2],[1, 2, 0, 1]])
newarr = arr.reshape(4, 3)
print ("\nOriginal array:\n", arr)
print ("Reshaped array[4,3]:\n", newarr)
```

```python
#Flatten array
flarr= arr.flatten()
print ("\nOriginal array:\n", arr)
print ("Fattened array:\n", flarr)
```

```python
# Printing array dimensions (axes)
print("\nNo. of dimensions: ", arr.ndim)
```

```python
# Printing shape of array
print("\nShape of array: ", arr.shape)
```

```python
# Printing size (total number of elements) of array
print("\nSize of array: ", arr.size)
```

```python
# Printing type of elements in array
print("\nArray stores elements of type: ", arr.dtype)
```
