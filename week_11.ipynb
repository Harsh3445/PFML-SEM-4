To apply Python concepts into a general-purpose Engineering feature

Description:

Numerical data such as continuous, highly skewed data is frequently seen in data analysis.
Sometimes analysis becomes effortless on conversion from continuous to discrete data.
There are many ways in which conversion can be done, one such way is by using Pandas'
integrated cut-function.
Pandas' cut function is a distinguished way of converting numerical continuous data into
categorical data.

Syntax: pandas.cut(x, bins, right=True, labels=None, retbins=False, precision=3,
include_lowest=False, duplicates='raise', ordered=True)
Parameters:
. x: Input array. Need to be 1-dimensional.
. bins: Denotes the bin boundaries for segmentation
· right: Denotes whether rightmost edge of bins should be included or not. Boolean
type of value. Default value is True.
· labels: Defines labels for returned segmented bins. Array or boolean
Return Value: Returns a Categorical series/numpy array/IntervalIndex
Example: Let's say we have an array 'Age' of 15 random numbers from 1 to 100 and we
wish to separate data into 4 bins of categories -

'Baby/Toddler' :- 0 to 3 years
'Child' :- 4 to 17 years
'Adult' :- 18 to 63 years

```python
import pandas as pd
import numpy as np

# Creating a dummy DataFrame of 15 numbers randomly
# ranging from 1-100 for age
df = pd.DataFrame({'Age': [42, 15, 67, 55, 1, 29, 75, 59, 4,
                          10, 15, 38, 22, 77]})

# Printing DataFrame Before sorting Continuous to Categories
print("Before:\n")
print(df)

# A column of name 'Label' is created in DataFrame
# Categorizing Age into 4 Categories
# Baby/Toddler: (0,3], 0 is excluded & 3 is included
# Child: (3,17], 3 is excluded & 17 is included
# Adult: (17,63], 17 is excluded & 63 is included
# Elderly: (63,100], 63 is excluded & 100 is included (assuming 100 as an upper limit)

bins = [0, 3, 17, 63, 100] # Defining bin boundaries
labels = ['Baby/Toddler', 'Child', 'Adult', 'Elderly'] # Defining labels for the bins

# Apply pd.cut to categorize the 'Age' column
df['Label'] = pd.cut(x=df['Age'], bins=bins, labels=labels, right=True, include_lowest=True)

# Printing DataFrame after sorting Continuous to Categories
print("\nAfter:\n")
print(df)

# Check the number of values in each bin
print("\nCategories:\n")
print(df['Label'].value_counts())
```

To apply Python concepts into a general-purpose Engineering feature

Description:

Numerical data such as continuous, highly skewed data is frequently seen in data analysis.
Sometimes analysis becomes effortless on conversion from continuous to discrete data.
There are many ways in which conversion can be done, one such way is by using Pandas'
integrated cut-function.
Pandas' cut function is a distinguished way of converting numerical continuous data into
categorical data.

Syntax: pandas.cut(x, bins, right=True, labels=None, retbins=False, precision=3,
include_lowest=False, duplicates='raise', ordered=True)
Parameters:
. x: Input array. Need to be 1-dimensional.
. bins: Denotes the bin boundaries for segmentation
. right: Denotes whether rightmost edge of bins should be included or not. Boolean
type of value. Default value is True.
. labels: Defines labels for returned segmented bins. Array or boolean
Return Value: Returns a Categorical series/numpy array/IntervalIndex

Example: Let's say we have an array 'Height' of 12 random people starting from 150cm to
180cm and we wish to separate data into 3 bins of categories.

'Short' :- greater than 150cm upto 157cm
'Average' :- greater than 157cm upto 170cm
'Tall' :- greater than 170cm upto 180cm

```python
# Importing pandas and numpy libraries
import pandas as pd
import numpy as np

# Creating a dummy DataFrame of 12 numbers randomly
# ranging from 150-180 for height
df = pd.DataFrame({'Height': [150.4, 157.6, 170, 176, 164.2, 155,
                             159.2, 175, 162.4, 176, 153, 170.9]})

# Printing DataFrame Before Sorting Continuous to Categories
print("Before:\n")
print(df)

# A column of name 'Label' is created in DataFrame
# Categorizing Height into 3 Categories
# Short: (150,157], greater than 150cm and upto 157cm
# Average: (157,170], greater than 157cm and upto 170cm
# Tall: (170,180], greater than 170cm and upto 180cm
bins = [150, 157, 170, 180] # Defining bin boundaries
labels = ['Short', 'Average', 'Tall'] # Defining labels for the bins

# Apply pd.cut to categorize the 'Height' column
df['Label'] = pd.cut(x=df['Height'], bins=bins, labels=labels, right=True, include_lowest=False)

# Printing DataFrame After Sorting Continuous to Categories
print("\nAfter:\n")
print(df)

# Check the number of values in each bin
print("\nCategories:\n")
print(df['Label'].value_counts())
```
