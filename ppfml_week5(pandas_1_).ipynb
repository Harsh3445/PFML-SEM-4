```python
import pandas as pd

# Define the Series and create the DataFrame 'result'
emp = pd.Series(["parker", "john", "smith", "williams"])
id = pd.Series([102, 203, 3098, 114])
frame = {'emp': emp, "id": id}
result = pd.DataFrame(frame)

# Define the DataFrame 'd2'
d2 = pd.DataFrame([['Dale', 123], ['Mark', 143]], columns = ['Emp', 'ID'])

# Concatenate the two DataFrames
concatenated_df = pd.concat([result, d2])
print(concatenated_df)
```
