```python
#Open a text file in 'Write' mode to write the given lines
f=open("textfile.txt", 'w')
f.write("first line\n")
f.write("Second line\n")
f.write("third line\n")
f.close()
#Open a text file in 'Read' mode to read and print the available lines - readline function
k=open("textfile.txt", 'r')
print(k.readline())
print(k.readline())
print(k.readline())
k.close()
```

```python
#Open a text file in 'Read' mode to read and print the available lines - using thee readlines function
k=open("textfile.txt", 'r')
print(k.readlines())
```

```python
# Open the text file in 'Append' mode to add new lines
f_append = open("textfile.txt", 'a')
f_append.write("\nThis is a new line appended.\n")
f_append.write("And another new line has been added.")
f_append.close()

# Now, open the file in 'Read' mode again to see the updated content
print("Content after appending:")
with open("textfile.txt", 'r') as f_read_appended:
    print(f_read_appended.read())
```

```python
import json

# JSON string
x='{"name":"John", "age":30, "city":"New York"}'

# Convert JSON string to Python dictionary
y=json.loads(x)
print(y["age"])
```
